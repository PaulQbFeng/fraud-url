```python
import pandas as pd
import tensorflow as tf
tf.__version__
```

```python
tf.enable_eager_execution()
```

```python
file_path = './dataset_no_dupli_no_http_only_url.csv'
```

```python
df = pd.read_csv(file_path, sep=',')
```

```python
CSV_COLUMNS = list(df.columns)
LABEL_COLUMN = 'isHiddenFraudulent'
# DEFAULTS = [['none'], ['Unknown'], ['True'], [0.0], ['Unknown'], ['Unknown'], ['Unknown'],['None']]

DEFAULTS = [tf.constant(['url'], dtype=tf.string),    # Column 0
                   tf.constant(['False'], dtype=tf.string)]   # Column 1
def decode_csv(row):
    columns = tf.decode_csv(row, record_defaults=DEFAULTS)
    features = dict(zip(CSV_COLUMNS, columns))
    label = features.pop(LABEL_COLUMN)
    return features, label

# Use prefix to create file path
file_path = './dataset_no_dupli_no_http_only_url.csv'

# Create dataset from file list
dataset = tf.data.TextLineDataset(file_path).map(decode_csv).shuffle(4).batch(4)
```

```python
features, label = dataset.make_one_shot_iterator().get_next()
```
